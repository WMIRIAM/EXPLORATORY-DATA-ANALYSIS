```python
#sub-category popularity based on region
plt.figure(figsize= (20, 8))
sns.countplot(x='Sub-Category', hue='Region', data=df)
plt.title('Sub-category popularity regionwise')
plt.show()
```

```python
#from the table, the western region is leading in orders in almost all subcategories.as such, people from the wetsern region order more from the store.
```

```python
#to better understand the store preformance, i will calculate the profit base margin for the sub-categories
df.head(5)
```

```python
df['Cost']=df['Sales']- df['Profit']
```

```python
df['Profitbasemargin']=(df['Profit']/ df['Cost'])*100
df.head()
```

```python
#top 10 products wrt to profitbase margin
df.sort_values(['Profitbasemargin', 'Product Name'], ascending=False).groupby('Profitbasemargin').head(10)
```

retailers that sold phones, binders paper had percent profit.

CUSTOMER LEVEL ANALYSIS

```python
#number of customers that purchased from the store
df['Customer ID'].nunique()
```

```python
#top 10 customers ordering from the store
df_top10=df['Customer Name'].value_counts().head(10)
df_top10
```

```python
#segment leading in purchases
plt.figure(figsize=(15, 8))
sns.countplot(x='Segment', hue='Region', data=df)
plt.title('segment orders by region')
plt.ylabel('Number of orders')
plt.show()
```

the consumer segment leads in the number of purchases, followed by the corporate and the home office segments respectively.  

```python
#top 20 customers who benefitted the store. These are the customers from who the store generated the highest profits. 
top20=df.sort_values('Profit', ascending=False).head(20)
fig=plt.figure(figsize=(20, 8))
ax=fig.add_subplot(111)
sns.barplot(x='Customer Name', y='Profit', palette='Set1', data=top20, ax=ax)
plt.title('Top twenty customers in profit')
plt.tight_layout()
plt.show()
```

```python
#top 20 customers who benefitted the store and by state

top20=df.sort_values('Profit', ascending=False).head(20)
fig=plt.figure(figsize=(20, 8))
ax=fig.add_subplot(111)
sns.barplot(x='Customer Name', y='Profit', hue='State', palette='Set1', data=top20, ax=ax)
plt.title('Top twenty customers in profit by State')
plt.tight_layout()
plt.show()
#Majority of the profitable customers come from the newyork and michigan states.
```
